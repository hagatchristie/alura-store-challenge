# 6. Criando visualizações com Matplotilb

```python
from numpy import rot90
import matplotlib.pyplot as plt

# Criando uma série com o faturamento de cada loja
faturamento_por_loja = pd.Series({
    'Loja 1': fat_total,
    'Loja 2': fat_total2,
    'Loja 3': fat_total3,
    'Loja 4': fat_total4
})

# Criar um gráfico de linhas
plt.figure(figsize=(10, 6))
plt.plot(faturamento_por_loja.index, faturamento_por_loja.values, marker='o', linestyle='--', color='red')
plt.xlabel("Loja")
plt.ylabel("Faturamento Total")
plt.title("Faturamento Total por Loja")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np # Import numpy for arange

# criando uma série com o faturamento de cada loja
produto_mais_menos_por_loja = pd.Series({
    'Loja 1': max_min_loja,
    'Loja 2': max_min_loja2,
    'Loja 3': max_min_loja3,
    'Loja 4': max_min_loja4
})

# extraindo dados para o gráfico
stores = produto_mais_menos_por_loja.index
max_products_names = [item[0] for item in produto_mais_menos_por_loja.values]
max_products_counts = [item[1] for item in produto_mais_menos_por_loja.values]
min_products_names = [item[2] for item in produto_mais_menos_por_loja.values]
min_products_counts = [item[3] for item in produto_mais_menos_por_loja.values]

# criando uma figura com largura e altura
plt.figure(figsize=(22, 15))

# definindo tamanho do gráfico e posição
bar_height = 0.30 # Height of the bars
ind = np.arange(len(stores)) # local de Y para colocar lojas com contagem

# Plot bars for Most Sold Products
# Each store will have its 'Mais Vendido' bar slightly above its y-position
plt.barh(ind + bar_height/2, max_products_counts, height=bar_height, color='skyblue', label='Mais Vendido')

# Plot bars for Least Sold Products
# Each store will have its 'Menos Vendido' bar slightly below its y-position
plt.barh(ind - bar_height/2, min_products_counts, height=bar_height, color='salmon', label='Menos Vendido')

plt.xlabel("Quantidade Vendida")
plt.ylabel("Loja")
plt.title("Produtos Mais e Menos Vendidos por Loja")

# Set y-axis ticks and labels to show store names
plt.yticks(ind, stores)

# Add annotations to the bars for product names and counts
for i, (max_name, max_count) in enumerate(zip(max_products_names, max_products_counts)):
    plt.text(max_count, ind[i] + bar_height/2, f' {max_name} ({max_count})', va='center')

for i, (min_name, min_count) in enumerate(zip(min_products_names, min_products_counts)):
    plt.text(min_count, ind[i] - bar_height/2, f' {min_name} ({min_count})', va='center')

plt.legend()
plt.tight_layout()
plt.show()
```

# **Análise para decisão de Venda de Loja**

## **Introdução**

O Sr. João possui quatro lojas em operação e necessita identificar qual delas deve ser vendida. Para apoiar essa decisão, foi realizada uma análise considerando os principais indicadores de desempenho: faturamento total, categorias de produtos mais e menos vendidas, avaliações dos clientes, produtos mais e menos procurados e frete médio de cada loja. O objetivo deste relatório é apresentar os resultados e justificar, de forma clara, qual loja apresenta o menor potencial estratégico para o negócio.

---

## **Análise dos Indicadores**

### **1. Faturamento total**

* **Loja 1:** maior faturamento
* **Loja 2:** segundo maior faturamento
* **Loja 3:** terceiro maior faturamento
* **Loja 4:** menor faturamento

### **2. Avaliação média dos clientes**

* **Loja 3:** melhor avaliação
* **Loja 2:** segunda melhor avaliação
* **Loja 4:** avaliação intermediária
* **Loja 1:** pior avaliação

### **3. Frete médio**

* **Loja 4:** menor frete
* **Loja 2:** segundo menor frete
* **Loja 3:** terceiro menor frete
* **Loja 1:** maior frete

### **4. Categorias de produtos mais e menos vendidas**

A análise mostrou que cada loja possui desempenho distinto por categoria, porém essas diferenças não compensam o baixo faturamento da Loja 4.

---

## **Conclusão**

Com base na análise dos dados de faturamento, avaliações, frete e desempenho por categoria, **a loja mais indicada para venda é a Loja 4**, pois:

* É a loja com **menor faturamento**;
* Possui **avaliação intermediária**, sem destaque positivo.
* Apesar do **frete baixo**, essa vantagem isolada não supera o desempenho inferior nos demais indicadores.

Dessa forma, a **Loja 4 apresenta o menor potencial estratégico**, sendo a escolha mais adequada para venda.
